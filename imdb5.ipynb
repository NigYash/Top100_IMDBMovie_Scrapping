```python
#import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
Books=[]
```

```python
for i in range(0,100):
    url = 'https://www.imdb.com/chart/top'
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'}
    response = requests.get(url,headers=headers)
    soup = bs(response.content, 'html.parser')
    #print(soup.prettify())
    main_box=soup.find_all('li',{'class':"ipc-metadata-list-summary-item sc-1364e729-0 caNpAE cli-parent"})
    main_box[i]
    product_link='https://www.imdb.com'+main_box[i].a['href']
    product=requests.get(product_link,headers=headers)
    product_link
    product_page=bs(product.text,'html.parser')
    #print(product_page.prettify())

    
    # print(product_page.prettify())
    movie_name=product_page.find('span', class_="hero__primary-text").text
    print(movie_name)
    ratings=product_page.find('span', class_='sc-bde20123-1 cMEQkK').text
    print(ratings)
    Director_name=product_page.find('a',class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link").text
    print(Director_name)
    top =product_page.find('div',class_="ipc-sub-grid ipc-sub-grid--page-span-2 ipc-sub-grid--wraps-at-above-l ipc-shoveler__grid")
    Top_Casts = [i.find(class_='sc-bfec09a1-1 gCQkeh').text for i in top]
    print(Top_Casts)
    Books.append([movie_name,ratings,Director_name, Top_Casts])
        
```

```python
df = pd.DataFrame(Books, columns=['Movie Name','Movie Rating', 'Director Name', 'Top Casts'])
df.to_csv('Top100Movies.csv')
```
